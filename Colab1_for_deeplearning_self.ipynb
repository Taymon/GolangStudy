##Imports

Let's start with importing the necessary libraries and frameworks. We wll be importing TensorFlow, Numpy, and Keras from TensorFlow. Keras is a framework defining neural network as a set of Sequential layers. 

Here they are:

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

##Define and Compile the Neural Network

Below we are create the simplest possible model of NN. It has a single layer with single neuron, and the input shap to it is just a single value as well.

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

When we define a NN, we have to specify 2 functions, a loss function and an optimizer. These functions encapsulate the Maths behind for Machine Learning. Basically, the loss function measures that how well or badly the model predict. The optimizer function is responsible to try to minimize the loss based on how the loss function went. This will repeat for a number of EPOCHS. 

In this case we will be using "Mean Square Error" for loss function and Stochastic Gradient Desceant (SGD) for optimizer.


```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

##Providing the Data

We will provide training data which in this case is 2 sets of arrays. By looking at the numbers, you cam deduce that the relationship is in fact y = 2x - 1. 

We will crate the 2 arrays using python library called Numpy.

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

##Training the Neural Network model

What we mean by training here is where the NN algorithm will 'learn' the relationship between the Xs and Ys. This happens when we call **model.fit** and it will use the optimizer and loss functions and after a number of times (EPOCHS), the NN will get to able to predict.

```
model.fit(xs, ys, epochs=500)
```

After training the NN model, we are now able to test by trying to predict a random value, let's say 10. See if it is about 19 then our model is doing great.

```
print(model.predict([10.0]))
```
